```python
# 1. Top Scorers by Position
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Pos', y='PTS')
plt.title('Points Distribution by Player Position')
plt.show()
```

```python
1.Identifies which positions contribute the most to scoring.
```

```python
# 2. Team-wise Average Performance
team_avg = df.groupby('Team')[['PTS', 'AST']].mean().sort_values(by='PTS', ascending=False)
plt.figure(figsize=(12, 5))
team_avg[['PTS', 'AST']].plot(kind='bar', figsize=(12, 5))
plt.title('Average Team Performance (Points & Assists)')
plt.ylabel('Average Value')
plt.xticks(rotation=90)
plt.show()
```

```python
2.  Analyzes how different teams perform on average.
```

```python
# 3. Three-Point Shooting Efficiency by Position
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Pos', y='3P%')
plt.title('Three-Point Percentage by Position')
plt.show()
```

```python
3.  Highlights which positions are best at three-point shooting
```

```python
# 4. Height vs. Points Scored
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Ht', y='PTS', alpha=0.5)
plt.title('Height vs Points Scored')
plt.xlabel('Height')
plt.ylabel('Points Per Game')
plt.show()
```

```python
4.  Examines if taller players score more points.
```

```python
# 5. Assist Leaders and Their Impact
assist_leaders = df.sort_values(by='AST', ascending=False).head(10)
plt.figure(figsize=(12, 5))
sns.barplot(data=assist_leaders, x='Player', y='AST', color='blue', label='Assists')
sns.barplot(data=assist_leaders, x='Player', y='PTS', color='orange', label='Points')
plt.xticks(rotation=90)
plt.title('Top Assist Leaders and Their Scoring Contribution')
plt.legend()
plt.show()
```

```python
5. Identifies players who contribute most in assists and scoring.
```
